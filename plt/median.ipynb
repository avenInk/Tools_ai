## primera pregunta

```python
x = np.linspace(1, 50, 50, dtype='int') #declaramos la cantidad de n

f = 8 * x ** 2  #Insertion sort    f(x) Verde
g = 64 * x * np.log2(x) #Merge sort  g(x) Rojo

plt.plot(x, f, 'g')
plt.plot(x, g, 'r')
plt.grid()
plt.show()

```

```python
data = {'x': x, 'f': f, 'g': g}
df = pd.DataFrame(data)
df = df.set_index('x')

df[f>g]
```

```python
df
```

## Segunda pregunta

```python
x = np.linspace(0, 17, 18, dtype='int') #declaramos la cantidad de n

f = 100 * (x ** 2)  #        f(x) Verde
g = 2 ** x #                 g(x) Rojo

plt.plot(x, f, 'g')
plt.plot(x, g, 'r')
plt.grid()
plt.show()
x
```

```python
data = {'x': x, 'f': f, 'g': g}
df = pd.DataFrame(data)
df = df.set_index('x')

df[f<g]
```
