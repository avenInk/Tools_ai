```python
import timeit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
import seaborn as sns
```

```python
#cargamos el dataset
X, y = datasets.load_diabetes(return_X_y=True)
X.shape
newy = y.copy()
```

```python
#cambiamos la dimensi√≥n del dataset
raw = X[:, None, 2]
raw.shape
```

```python
max_value = max(raw)
min_value = min(raw)

scaller = ((2*raw - min_value - max_value)/(max_value - min_value))
```

```python
fig, axis = plt.subplots(2, sharex=True)

axis[0].hist(scaller)
axis[1].hist(raw)
```

```python
def re_n():
    linear_model.LinearRegression().fit(raw, newy)
def re_s():
    linear_model.LinearRegression().fit(scaller, newy)
```

```python
raw_time = timeit.timeit(re_n, number=100)
scaller_time = timeit.timeit(re_s, number=100)

print(f"Raw time: {raw_time}")
print(f"climbing time: {scaller_time}")
```
