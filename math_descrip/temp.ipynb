```python
import matplotlib.pyplot as plt
import numpy as np
import time
from IPython.display import clear_output
```

```python
%matplotlib inline

repeat = 100

values = np.linspace(0, 3, repeat)

for i in range(repeat):

    res = np.sqrt(np.abs(values[i]**2 - 3**2))
    # x = values[i]    y = res

    x_first = np.array([0, values[i]])
    x_second = np.array([0, res])

    y_fist = np.array([0, res])
    y_second = np.array([0, values[i]])

    plt.plot([0, 0], [0, 6], alpha=0)
    plt.plot([0, 6], [0, 0], alpha=0)

    plt.plot(x_first+x_second, y_fist + y_second)
    plt.plot(x_first, y_fist) #referencia
    plt.plot(x_second, y_second) #referencia

    plt.grid()
    
    plt.show()
    plt.clf()
    plt.close('all')
    clear_output(wait=True)

```
