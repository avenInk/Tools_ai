```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['Mexico', 'Japan', 'Colombia', 'Alemania']
population = [500, 300, 1000, 809]
```

```python
plt.bar(country, population, color=["red", "blue"], width = 0.6)
plt.show()
```

```python
plt.bar(country, population, color=["red", "blue"], width = 0.6)
plt.xticks(np.arange(4), ("India",  "Japón",  "México", "Alemania"), rotation=45)
plt.show()
```

```python
plt.barh(country, population)
```
