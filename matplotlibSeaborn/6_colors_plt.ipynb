```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5, 20)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize= (4.5,4.5))

ax.plot(x, x+1, 'r--')
ax.plot(x, x+2, 'b-')
ax.plot(x, x+3, 'y:')
ax.plot(x, x+4, 'D')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(4.5, 4.5))
ax.plot(x, x+1, color='green', linewidth = 2, linestyle = '--', marker='o', markersize=8)
ax.plot(x, x+2, color='yellow', linewidth = 3, linestyle = ':', marker='D', markersize=4)
ax.plot(x, x+3, color='#66FF66', linewidth = 2, linestyle = 'dashed')
ax.plot(x, x+4, color='#CC22CC', alpha = 0.3, linewidth = 8, linestyle = '-')
```
