```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 30)
y = np.sin(x)
```

## Forma tradicional


```python
fig = plt.figure()
axes1 = fig.add_axes([0.3, 0.3, .2, .2])
axes1.plot(x, y)
fig.show()
```

## Forma subplots


```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'rx--')
axes[1].plot(y, x, 'b:')
fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax2.plot(y, x)
fig.show()
```

```python
fig, axes = plt.subplots(2, 4)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x), 'y')
axes[0, 2].plot(x, np.tan(x), 'r')
axes[0, 3].plot(x, np.cos(x)**2)
fig.show()
```

```python
fig, [(ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8)] = plt.subplots(2, 4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x)**2)
ax5.plot(y,x, ':')
ax6.plot(np.sin(y), x**3)
ax7.plot(np.cos(y), np.tan(x), 'y--')
ax8.plot(np.tan(x)**2, np.sin(y))
fig.tight_layout() # para que sea más estilizado
fig.show()
```
