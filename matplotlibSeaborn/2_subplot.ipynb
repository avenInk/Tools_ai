```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10, 10, dtype='int8')
y = x**2 
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'ro--')
plt.plot(y,x, 'b:')
plt.subplot(2,1,2)
plt.pie(x)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'ro--')
plt.plot(y,x, 'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```
