```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 20)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes.plot(x,y, 'o--')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.5, 0.9])
axes2 = fig.add_axes([0.14, 0.5, 0.3, 0.3])
axes.plot(x,y, 'o--')
axes2.plot(x,y, 'b:')
```
