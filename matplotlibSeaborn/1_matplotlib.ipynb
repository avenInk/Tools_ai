```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,10, dtype='int8')

y = x**2
```

```python
plt.plot(x,y, 'r') # r, y, b 

plt.show()
```

```python
plt.plot(x,y, 'ro') # para poner puntitos u.u 
plt.show()
```

```python
plt.plot(np.random.randint(1,150, 500), np.random.randint(1,150,500), 'rx')

plt.show()
```

```python
plt.plot(x,y, 'rs-')
plt.show()
```

```python
plt.plot(x,y, 'yD:')
plt.show()
```

```python
plt.hist(y)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y) # no veo mucha diferencia entre esto y hacerlo en el plot, supongo que ahorras cdigo.
plt.show() 
```

```python
plt.boxplot(x)
plt.show()
```
