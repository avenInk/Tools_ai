```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15,
             cumulative=False, hue='sex', stat='percent')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False,
            hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', shade=True, bw_adjust=1)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='hist', cumulative=False, multiple='stack')
plt.show()
```
