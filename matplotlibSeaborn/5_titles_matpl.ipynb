```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 20)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(4,3))
ax[0].plot(x, y, 'r', label='$sin(x)$')
ax[0].legend() # Para que salga el laben que definimos
ax[0].set_title('Sin X')
ax[1].plot(y, x)
ax[1].set_title('Sin de X en Y')

ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[1].set_ylabel('Y')
ax[1].set_xlabel('X')

fig.tight_layout() #para acommodar todo

fig.show()
```
