```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1,50, 1000)
```

## Histograma

```python
plt.hist(data, bins = 40, width = 1)
plt.show()
```

## Boxplot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Scatter

```python
x = np.random.rand(50)
y = np.random.rand(50)

area = np.random.rand(50)*30**2
```

```python
plt.scatter(x, y, s=area, alpha=0.6, marker='p')
plt.show()
```
