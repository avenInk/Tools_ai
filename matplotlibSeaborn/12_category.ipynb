```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(4)
```

```python
plt.figure(figsize=[5,5])
sns.countplot(data=tips, x='day', hue='sex')
plt.show()
```

```python
plt.figure(figsize=[5,5])
sns.stripplot(data=tips, x='day', y='tip', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=[5,5])
sns.swarmplot(data=tips, x='day', y='tip', hue='sex', dodge=True)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color='0', marker='<')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', palette='pastel', split=True)
plt.show()
```

```python
sns.violinplot(data=tips, y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex', kind='box', col='time')
```
