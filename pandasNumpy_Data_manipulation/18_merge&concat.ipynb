# Merge

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", 'D3']
})
```

```python
izq.merge(der, on='key') # on= especifica la columna que tienen en comun los dos dataframe
```

```python
izq = pd.DataFrame({'key_1':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", "D3"]
})

izq.merge(der, on='key')     # Esto genera un error porque en este dataframe ya no hay key clonado.
```

```python
izq = pd.DataFrame({'key_1':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", "D3"]
})

izq.merge(der, left_on='key_1', right_on='key')  # ahora especificamos la columna de izquierda y de derecha
```

En el codigo de abajo no se ponen los valores del index 3 porque la relación está en key, pero en este caso hemos borrado k3 en uno de los dataframe, por lo cual el k3 de izq no se relaciona con el nan de der en el index 3, por lo cual el index 3 no tiene relación alguna entre dataframes y no se muestra en el merge.

```python
izq = pd.DataFrame({'key_1':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2', np.nan],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", "D3"]
})

izq.merge(der, left_on='key_1', right_on='key') 
```

```python
izq = pd.DataFrame({'key_1':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2', np.nan],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", "D3"]
})

izq.merge(der, left_on='key_1', right_on='key', how='left') 
#ahora sólo se muestran los de la izquierda aunque no haya relación en el index 3
#a la derecha no sale nada porque no hay relación, forzamos la aparición de los valores
#de la izquierda.
```

```python
izq = pd.DataFrame({'key_1':['k0','k1','k2','k3'],
                    'A':["A0", "A1", "A2", "A3"],
                    'B':["B0", "B1", "B2", "B3"]
})

der = pd.DataFrame({'key':['k0','k1','k2', np.nan],
                    'C':["C0", "C1", "C2", "C3"],
                    'D':["D0", "D1", "D2", "D3"]
})

izq.merge(der, left_on='key_1', right_on='key', how='right') 
#ahora forzamos los valores de la derecha.
```
