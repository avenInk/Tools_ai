```python
import pandas as pd
```

```python
df = pd.read_csv('./resources/bestsellers with categories.csv')
```

```python
df[df['Year'] == 2016].shape # para saber cuantas entradas hay del año 2016
```

```python
df.groupby('Year').count()
```

```python
df.groupby('Author').count() # .max  .sum .min ()
```

```python
df.groupby('Author').count().reset_index() # se usa reset index porque AUTHOR era el indice y regresamos a los números
```

```python
df.groupby("Author").agg(["max", "min"]) #".agg" para decirle que quiero la agrupación con funciones de agregación, en concreto max y min.
```

```python
df.groupby("Author").agg({"Reviews":['min', 'max'],
                            "Year":['min', 'max'],
                            'User Rating':'sum'}).reset_index()
```

```python
data = df.groupby(['Author', 'Year']).count() #guardamos la agrupación del autor y el año 
data[data['Name']>1] #en la agrupación filtramos sólo cuando en un año un autor haya sacado más de un libro.
```
