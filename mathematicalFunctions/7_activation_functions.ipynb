```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)
```

### Lineal


```python

def f(x):
    return x
plt.figure(figsize=(4,4))
plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Sigmoid

```python
def f(x):
    return 1/(1+np.e**-x)

plt.figure(figsize=(4,4))
plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Heaviside

```python
def f(x):
    Y = np.zeros(len(x))
    for index, unit in enumerate(x):
        if unit > 0:
            Y[index] = 1
    return Y

plt.figure(figsize=(4,4))
plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Relu

```python
def f(x):
    return np.maximum(0,x)

plt.figure(figsize=(4,4))
plt.plot(x, f(x))
plt.grid()
plt.show()
```

### Hyperbolic tan

```python
def f(x):
    return np.tanh(x)

plt.figure(figsize=(4,4))
plt.plot(x, f(x))
plt.grid()
plt.show()
```
