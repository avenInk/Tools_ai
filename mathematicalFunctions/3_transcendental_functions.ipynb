```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data
x = np.linspace(-10, 10, 100)
```

## trigonometric functions


```python
def f(x):
    return np.cos(x)


y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
fig.show()
```

## exp functions


```python
def f(x):
    return np.exp(x)


y = f(x)

plt.plot(x, y)
plt.show()
```

## Log


```python
def f(x):
    return np.log2(x)


_x = np.linspace(0.001, 10, 1000)

y = f(_x)

plt.plot(_x, y)
plt.show()
```
