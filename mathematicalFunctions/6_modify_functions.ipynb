```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 200)

f = lambda x: np.sin(x)

y = f(x)
plt.figure(figsize=(3,3))
plt.plot(x, y)
plt.grid()
plt.show()
```

### up and down

```python
y = f(x)+0.22 #up 
y_ = f(x)-0.5 #down
```

```python
fig, (ax, ax2) = plt.subplots(2, figsize=(3,3))
ax.axhline(y=-0.73, color='r')
ax2.axhline(y=-0.73, color='r')

ax.plot(x, y)
ax2.plot(x, y_)
ax.set_title('up')
ax2.set_title('down')
plt.tight_layout()

ax.grid()
ax2.grid()
fig.show()
```

### Side by side

```python
g = lambda x: x**2

y = g(x+2) #up 
y_ = g(x-2) #down
```

```python
fig, (ax, ax2) = plt.subplots(2, figsize=(3,3))
ax.axvline(x=-0, color='r')
ax2.axvline(x=-0, color='r')

ax.plot(x, y)
ax2.plot(x, y_)
ax.set_title('left')
ax2.set_title('right')
plt.tight_layout()

ax.grid()
ax2.grid()
fig.show()
```

### stretch and compress

```python
y = f(x/3) # stretch
y_ = f(x*3)-0.22 #compress
```

```python
fig, (ax, ax2) = plt.subplots(2, figsize=(4,4))
ax.axhline(y=-0.73, color='r')
ax2.axhline(y=-0.73, color='r')

ax.plot(x, y)
ax2.plot(x, y_)
ax.set_title('stretch')
ax2.set_title('compress')
plt.tight_layout()

ax.grid()
ax2.grid()
fig.show()
```

#### vertical axis

```python
y = 2 * f(x) # stretch
y_ = 1/2 * f(x)-0.22 #compress
```

```python
fig, (ax, ax2) = plt.subplots(2, figsize=(4,4))

ax.axhline(y=-0.25, color='r')
ax2.axhline(y=0.25, color='r')
ax.axhline(y=-0.75, color='r')
ax2.axhline(y=-0.75, color='r')

ax.plot(x, y)
ax2.plot(x, y_)
ax.set_title('stretch')
ax2.set_title('compress')
plt.tight_layout()

ax.grid()
ax2.grid()
fig.show()
```

### reflections

```python
h = lambda x: x**3

y = h(x) #normal
y_ = -h(x) # X reflex
_y = h(-x) # Y reflex
```

```python
fig, ax = plt.subplots(2,2, figsize=(5,5))

ax[0,0].plot(x, y)

ax[1,0].plot(x, y_)
ax[1,1].plot(x, _y)

ax[0,1].plot(x, y)


ax[0,0].set_title('normal')
ax[1,0].set_title('X reflex -f(x)')

ax[0,1].set_title('normal')
ax[1,1].set_title('Y reflex f(-x)')
plt.tight_layout()


fig.show()
```
