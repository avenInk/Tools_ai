```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Range
x = np.linspace(-100, 100, 1000)
```

```python
def H(x):
    Y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x > 0:
            Y[idx] = 1
    return Y


y = H(x)

plt.plot(x, y)
plt.show()
```

## absolute functions

```python
def f(x):
    Y = x.copy()
    Y[Y<0] = Y[Y<0]*-1
    return Y

y = f(x)

plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.show()
```
